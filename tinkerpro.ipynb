## Function to find Data Using Studio_id

```python

def open_data_finder():
    new_data_finder_window = tk.Toplevel(root)
    new_data_finder_window.title("Find Data")
    
    # Studio Id
    studio_id_label = ttk.Label(new_data_finder_window, text="Studio Id:")
    studio_id_label.grid(row=0, column=0,padx=10,pady=5,sticky="w")
    studio_id_entry = ttk.Entry(new_data_finder_window)
    studio_id_entry.grid(row=0, column=1,padx=10,pady=5)
    
    # Find button
    find_button = ttk.Button(new_data_finder_window,text="Find Data",command=lambda:display_data(studio_id_entry.get()))
    find_button.grid(row = 1,column=0,columnspan=2,padx=10,pady=10)
    
    

def show_studio_data(data):
    studio_info_window = tk.Toplevel(root)
    studio_info_window.title("Studio Information")

    labels = ["Entry Id","Studio ID", "Studio Name", "Date of Arrival", "Cassette Id", "Wedding Name", "Wedding Date", "Pendrive"]
    
    for row_idx, row_data in enumerate(data):
        entry_frame = ttk.LabelFrame(studio_info_window, text=f"Entry {row_idx + 1}")
        entry_frame.grid(row=row_idx, column=0, padx=10, pady=5, sticky="w")

        for col_idx, label_text in enumerate(labels):
            label = ttk.Label(entry_frame, text=label_text + ":")
            label.grid(row=col_idx, column=0, padx=5, pady=5, sticky="w")

            value = ttk.Label(entry_frame, text=str(row_data[col_idx]))
            value.grid(row=col_idx, column=1, padx=5, pady=5)





def display_data(studio_id):
    try:
        connection = psycopg2.connect(
            user="postgres",  # Replace with your PostgreSQL username
            password="FocusOnMission22",  # Replace with your PostgreSQL password
            host="localhost",  # Replace with your PostgreSQL server host
            port="5432",  # Replace with your PostgreSQL port
            database="postgres"  # Replace with your PostgreSQL database name
        )
        
        cursor = connection.cursor()
        
        select_query = """
            SELECT * FROM register1 WHERE studio_id = %s
        """
        # Execute the SELECT statement with the provided studio_id
        cursor.execute(select_query, (studio_id,))
        studio_data = cursor.fetchall() ## to fetch all data to fetch one data use-> cursor.fetchone()
        
        # Close the cursor and the database connection
        cursor.close()
        connection.close()

        if studio_data:
            # If studio data is found, display it in a new window
            show_studio_data(studio_data)
        else:
            messagebox.showinfo("Studio Not Found", "Studio with the provided studio_id does not exist.")

    except (Exception, psycopg2.Error) as error:
        print("Error while connecting to PostgreSQL:", error)
        # Show an error message
        messagebox.showerror("Error", "An error occurred while retrieving the data.")
```
